## Using pandas and Jupyter Notebook 
- Jupyter Notebook (via Anaconda) is a web-based application that allows you to create and share documents containing live code, equations, visualizations, and narrative text 
- pandas is Python library that uses DataFrames

### DataFrame vs. Database
- Database:
	Optimized for transactions & storage, lives in storage/disk, 
	‘normalized data’ 
    
- DateFrame:
	Optimized for analytics, lives in memory, 
	‘de-normalized data’
    
#### Not all DataFrames are the same  
- Distributed (more than one computer)
	Apache Spark or Julia DB  
    
- Non-distributed (single node)
	pandas or
	Base R
